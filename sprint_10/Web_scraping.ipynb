```python
#usando beautifulsoup para hacer web scraping
import requests
from bs4 import BeautifulSoup


url = "http://quotes.toscrape.com"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
cita_auth = {}
for cita_div in soup.find_all('div', class_='quote'):
    text = cita_div.find('span', class_='text').text
    #obtener author 
    autor = cita_div.find('small', class_='author').text
    #almacenar en diccionario, si el author ya existe, se agrega la cita
    if autor in cita_auth:
        cita_auth[autor].append(text)
    else:
        cita_auth[autor] = [text]
```

**Ahora usando Selenium:**

```python
import os
from selenium import webdriver

# Establecer la variable de entorno PATH
os.environ['PATH'] += os.pathsep + '/usr/local/bin/'

# Inicializar el navegador
browser = webdriver.Chrome()
browser.get('http://quotes.toscrape.com')

citas_selenium = []
cita_elements = browser.find_elements_by_css_selector("div.quote")
for cita in cita_elements:
    try:
        texto = cita.find_element_by_css_selector("span.text").text
        autor = cita.find_element_by_css_selector("small.author").text
        citas_selenium.append((texto, autor))
    except Exception as e:
        print(f"Error: {e}.")
```

**Importante:** Aqui sólo se ha extraido la información de la primera página. Si quisieramos extraer todas las citas de todas las páginas, como BeautifulSoup sería complicado, ya que no puede interactuar con el navegador. Sin embargo, Selenium sí puede hacerlo. A continuación se muestra un ejemplo de cómo hacerlo.

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

# Configurar el navegador
driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')

# Navegar a la primera página de las citas
driver.get("http://quotes.toscrape.com")

citas = []

while True:
    # Extraer todas las citas de la página actual
    elementos_cita = driver.find_elements_by_css_selector("div.quote")
    for cita in elementos_cita:
        texto = cita.find_element_by_css_selector("span.text").text
        autor = cita.find_element_by_css_selector("small.author").text
        citas.append((texto, autor))

    # Intentar hacer clic en el enlace "Siguiente" para ir a la siguiente página
    try:
        boton_siguiente = driver.find_element_by_css_selector("li.next > a")
        boton_siguiente.click()
    except NoSuchElementException:
        # Si no podemos encontrar un botón "Siguiente", estamos en la última página
        break

# Cerrar el navegador después de hacer scraping
driver.quit()

# Aquí, "citas" es una lista de tuplas que contienen las citas y los autores
for cita in citas:
    print(cita)
```

```python
#ahora haremos el scraping de la pagina 2 con selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

URL = "https://www.bolsamadrid.es"

def configurar_navegador():
    # Configurar el navegador y retornar el driver
    return webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')

def obtener_informacion(driver):
    info_completa = []
    filas = driver.find_elements_by_css_selector("tr[role='row']")

    for fila in filas:
        try:
            info_fila = extraer_informacion_fila(fila)
            if info_fila:  # Si info_fila no está vacío
                info_completa.append(info_fila)
        except Exception as e:
            print(f"Error procesando fila: {e}")

    return info_completa

def extraer_informacion_fila(fila):
    info_fila = {}
    celdas = fila.find_elements_by_tag_name("td")

    if len(celdas) >= 4:
        info_fila["nombre"] = celdas[0].text
        info_fila["codigo"] = celdas[1].text
        info_fila["direccion"] = celdas[2].text.split("\n")[0]

        try:
            enlace = celdas[2].find_element_by_tag_name("a").get_attribute("href")
            info_fila["enlace"] = enlace
        except NoSuchElementException:
            info_fila["enlace"] = None

        info_fila["bolsa"] = celdas[3].text
        return info_fila

    print("Fila no contiene suficiente información.")
    return None 

# Ejecutando el código
driver = configurar_navegador()
driver.get(URL)

wait = WebDriverWait(driver, 20)

boton_cookies = wait.until(EC.element_to_be_clickable((By.ID, 'onetrust-accept-btn-handler')))
boton_cookies.click()

enlace = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), 'Miembros') and contains(@class, 'mover')]")))
enlace.click()

wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))

info_completa = obtener_informacion(driver)

for info in info_completa:
    print(info)

driver.quit()  # Asegurarse de cerrar el navegador al final
```

* Haré web scraping de la página de la IMDB, extraeré las 250 películas top

```python
#selenuim
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

url_imdb = 'https://www.imdb.com/'


def configurar_navegador():
    # Configurar el navegador y retornar el driver
    return webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')


driver = configurar_navegador()
driver.get(url_imdb)

wait = WebDriverWait(driver, 10)

# Esperar hasta que el elemento esté presente y hacer clic en él
menu = wait.until(
    EC.element_to_be_clickable((By.ID, "imdbHeader-navDrawerOpen"))
)
menu.click()

# Esperar hasta que el elemento esté visible y hacer clic en él
movies_dropdown = wait.until(
    EC.element_to_be_clickable((By.CLASS_NAME, "navlinkcat__itemTitle"))
)
movies_dropdown.click()

# Nuevamente click en el span que tiene como clase ipc-list-item__text
top_250_movies = wait.until(
    EC.element_to_be_clickable(
        (By.XPATH, "//span[contains(@class, 'ipc-list-item__text') and text()='Top 250 Movies']")
    )
)
top_250_movies.click()

películas_h3_elements = driver.find_elements_by_css_selector(
    'li h3.ipc-title__text'
)

# Iterar sobre la lista de elementos y obtener el texto de cada uno
nombres_peliculas = [element.text for element in películas_h3_elements]

# Ahora, nombres_peliculas es una lista con los nombres de las películas
print(nombres_peliculas)


driver.quit()
```

**Importante:** Scrapy es una biblioteca de Python muy poderosa para hacer scraping de páginas web, pero tiene un enfoque diferente en comparación con Selenium. Scrapy es mejor para recopilar datos de múltiples páginas en un sitio web de manera eficiente, mientras que Selenium es mejor para interactuar con páginas web de una manera que imita la interacción humana, especialmente en páginas con mucho JavaScript.

```python
import scrapy
from scrapy.crawler import CrawlerProcess
import warnings
warnings.filterwarnings('ignore')  # Ignorar warnings

# Lista global para recopilar los nombres de las películas
nombres_peliculas = []

class IMDBSpider(scrapy.Spider):
    name = 'imdb'
    start_urls = ['https://www.imdb.com/chart/top/?ref_=nv_mv_250']

    def parse(self, response):
        global nombres_peliculas  # Indicar que se utilizará la variable global
        # Extraer los nombres de las películas
        for pelicula in response.css('li h3.ipc-title__text::text'):
            nombres_peliculas.append(pelicula.get())  # Agregar cada nombre a la lista

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0',  # Especificar un user agent para evitar ser bloqueado
})

process.crawl(IMDBSpider)
process.start()  

# Convertir la lista de nombres de películas a un diccionario
diccionario_peliculas = {index: nombre for index, nombre in enumerate(nombres_peliculas, 1)}

print(diccionario_peliculas)
```
