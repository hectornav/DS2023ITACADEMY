```python
import pandas as pd
```

```python
#importamos los datos
df = pd.read_csv('Lligues_europees_UEFA.csv', encoding='ISO-8859-1', sep=';')
#convertimos a utf-8 para que no haya problemas con los acentos
df = df.apply(lambda x: x.astype(str).str.encode('utf-8').str.decode('utf-8'))
df
```

**EJERCICIO 1**: Descarga el dataset adjunto de datos oficiales de la UEFA y selecciona un atributo del conjunto de datos. Calcula el p-valor y di si rechaza la hipótesis nula cogiendo un alfa de 5%.

- Vamos a formular la hipótesis nula $H_0$. Para ello, seleccionaremos la variable GF (Goles a Favor).
  - $H_0$: El promedio de goles a favor de todos los equipos es 50.
- La hipótesis alternativa $H_1$ será:
  - $H_1$: El promedio de goles a favor de todos los equipos no es 50.

**Seleccionamos la prueba estadística a utilizar**: En este caso, como trabajamos con un promedio y una muestra, podemos usar la prueba t de una muestra.

**Tomamos una decisión**:
- Si el p-valor $< \alpha$ (0.05), rechazamos la hipótesis nula.
- Si el p-valor $> \alpha$ (0.05), no rechazamos la hipótesis nula.


```python
#cargamos la prueba t de student para una muestra
from scipy.stats import ttest_1samp
import numpy as np

#Datos de la columna GF
data = df['GF']

#convertimos los datos a float
data = data.astype(float)
#realizamos la prueba t de student para una muestra, de acuerdo a la hipótesis planteada, la media es 50
tset, pval = ttest_1samp(data, 50)
#imprimimos el p-value
print("p-value: ",np.round(pval,3))
```

**Finalmente, tomamos una decisión**:
- el p-value > 0.05, por lo que no rechazamos la hipótesis nula.

Esto significa que no hay suficiente evidencia para rechazar la hipótesis nula, por lo que no podemos decir que el promedio de goles a favor de todos los equipos es diferente de 50.

Realizaremos las pruebas t de una muestra para ambos atributos y luego compararemos los p-valores obtenidos con el nivel de significancia $\alpha = 0.05$.


```python
#iniciamos con la variable GA
data = df['GA']

#convertimos los datos a float
data = data.astype(float)

#realizamos la prueba t de student para una muestra, de acuerdo a la hipótesis planteada, la media es 40
tset, pval = ttest_1samp(data, 40)
#imprimimos el p-value
print("p-value: ",pval)
```

$ \alpha = 0.05 $

Para "Pts" (Puntos):

- $ p $-value:

Ambos p-values son extremadamente pequeños, lo que indica que los resultados son altamente significativos. Es decir, las diferencias observadas en los datos son muy poco probable que se deban al azar.


**EJERCICIO 3**: Continúa con el conjunto de datos adjunto y selecciona tres atributos del conjunto de datos. Calcula el p-valor y di si rechaza la hipótesis nula cogiendo un alfa de 5%.

Elegiremos "xG" (Expectativa de Goles), "xGA" (Expectativa de Goles En Contra) y "xGD" (Diferencia de Expectativa de Goles).
